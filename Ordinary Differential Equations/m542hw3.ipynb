```python
import numpy as np
import matplotlib.pyplot as plt
import fdscheme as fds

r = 28
start_time = 0
stop_time = 50
time_step = 0.01
sigma = 10
b=8/3
X0 = np.array([0, 
               1, 
               0]) # X0, Y0, Z0

def Lorenz_3D(t, X):
    x = X[0]
    y = X[1]
    z = X[2]
    x_dot = -sigma*x + sigma*y
    y_dot = -x*z + r*x - y
    z_dot = x*y - b*z
    system = np.array([x_dot, y_dot, z_dot])
    return system

t = fds.Grid(start_time, stop_time, time_step).create_grid()
solution = fds.Scheme(Lorenz_3D, X0, start_time, stop_time, time_step)
runge, name = solution.runge_kutta()

fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection ='3d')
 
# defining all 3 axes
z = runge[2,:]
x = runge[0,:]
y = runge[1,:]
 
# plotting
ax.plot3D(x, y, z)
ax.set_title('3D Lorenz System with $\\sigma=10$, $r=28$, $b=\\frac{8}{3}$; computed with ' +name, fontdict = {'fontsize' : 12})
#rot_animation = animation.FuncAnimation(fig, rotate, frames=np.arange(0,362,2),interval=100)
#rot_animation.save('rotation.gif', dpi=80, writer='imagemagick')
plt.show()
```
