```python
import numpy as np
import matplotlib.pyplot as plt
import fdscheme as fds
```

I decided to do 5D and 7D Lorenz. I attempted 9D Lorenz but it made all of the schemes we learned in class unstable. I plotted all parameters with respect to x. I used Runge-Kutta to calculate them

```python
r = 28
start_time = 0
stop_time = 50
time_step = 0.01
sigma = 10
b=8/3
X0 = np.array([np.random.uniform(0,1), 
               np.random.uniform(0,1), 
               np.random.uniform(0,1)]) # X0, Y0, Z0

def Lorenz_3D(t, X):
    x = X[0]
    y = X[1]
    z = X[2]
    x_dot = -sigma*x + sigma*y
    y_dot = -x*z + r*x - y
    z_dot = x*y - b*z
    system = np.array([x_dot, y_dot, z_dot])
    return system

t = fds.Grid(start_time, stop_time, time_step).create_grid()
solution = fds.Scheme(Lorenz_3D, X0, start_time, stop_time, time_step)
runge, name = solution.adams_moulton4()

fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection ='3d')
 
# defining all 3 axes
z = runge[2,:]
x = runge[0,:]
y = runge[1,:]
 
# plotting
ax.plot3D(x, y, z)
ax.set_title('3D Lorenz')
plt.show()
    
```

```python
r = 45
a = 1/np.sqrt(2)
d0 = (3**2+a**2)/(1+a**2)
X0 = np.array([np.random.uniform(0,1), 
               np.random.uniform(0,1), 
               np.random.uniform(0,1),
               np.random.uniform(0,1),
               np.random.uniform(0,1)])

def Lorenz_5D(t, X):
    x = X[0]
    y = X[1]
    z = X[2]
    y1 = X[3]
    z1 = X[4]
    x_dot = -sigma*x + sigma*y
    y_dot = -x*z + r*x - y
    z_dot = x*y - x*y1 - b*z
    y1_dot = x*z - 2*x*z1 - d0*y1
    z1_dot = 2*x*y1 - 4*b*z1
    system = np.array([x_dot, y_dot, z_dot, y1_dot, z1_dot])
    return system

t = fds.Grid(start_time, stop_time, time_step).create_grid()
solution = fds.Scheme(Lorenz_5D, X0, start_time, stop_time, time_step)
runge, name = solution.runge_kutta()

plt.figure(figsize = (7,20))
plt.subplot(4,1,1)
plt.plot(runge[0,:],runge[1,:])
plt.subplot(4,1,2)
plt.plot(runge[0,:],runge[2,:])
plt.subplot(4,1,3)
plt.plot(runge[0,:],runge[3,:])
plt.subplot(4,1,4)
plt.plot(runge[0,:],runge[4,:])
```
